```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def dcf(x):
    x = 4*x
    return (x+1) ** (1/4) * np.log(x+1) + 0.25

def capm(x):
    x = 4*x
    return x/(1+np.exp(-0.5*x))

def comps(x):
    x = 4*x
    return np.exp(0.4*x)-0.2-1/(0.5*np.sqrt(2*np.pi))*(np.e**(-0.5*(x/0.5)**2))

def ewma(x):
    x = 4*x
    return x**2 * np.sin(0.05*x)

def vix(x):
    x = 4*x
    return 1/x + 0.4*np.sin(x)

x = np.linspace(0, 1, 1000)
dcf_values = dcf(x)
capm_values = capm(x)
comps_values = comps(x)
ewma_values = ewma(x)
vix_values = vix(x)

plt.plot(x, dcf_values)
# plt.plot(x, capm_values)
plt.plot(x, comps_values)
plt.plot(x, ewma_values)
plt.plot(x, vix_values)
plt.ylim(0, 4)
plt.xlim(0, 1)
plt.ylabel('Return')
plt.xlabel('Normalized factors')
plt.legend(['DCF', 'COMPS', 'EWMA', 'VIX'])
plt.savefig('Nvidia model.png', dpi=400)
```
