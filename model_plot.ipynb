```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def dcf(x):
    x = 4*x
    return ((x+1) ** (1/4) * np.log(x+1) + 0.25)/8*100-25

def capm(x):
    x = 4*x
    return (x/(1+np.exp(-0.5*x)))/8*100-25

def comps(x):
    x = 4*x
    return (np.exp(0.4*x)-0.2-1/(0.5*np.sqrt(2*np.pi))*(np.e**(-0.5*(x/0.5)**2)))/8*100-25

def ewma(x):
    x = 4*x
    return (x**2 * np.sin(0.05*x))/8*100-20

def vix(x):
    x = 4*(x+0.04)
    return ((1/x + 0.4*np.sin(x))/8*100-25) * 0.8

x = np.linspace(0, 1, 10000)
dcf_values = dcf(x)
capm_values = capm(x)
comps_values = comps(x)
ewma_values = ewma(x)
vix_values = vix(x)

plt.plot(x, dcf_values)
# plt.plot(x, capm_values)
plt.plot(x, comps_values)
plt.plot(x, ewma_values)
plt.plot(x, vix_values)
plt.ylim(-20, 50)
plt.xlim(0, 1)
plt.ylabel('Return %')
plt.xlabel('Normalized variables (0-1)')
plt.legend(['DCF', 'COMPS', 'EWMA', 'VIX'], loc='upper right')
plt.title('Fig. 7: Relationship between each variable and annual return (Turing)', fontweight='bold', fontsize=11, pad=15)
plt.savefig('Nvidia model.png', dpi=400)
```
